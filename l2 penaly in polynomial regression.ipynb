```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function to produce an DataFrame with columns containing the powers of a given input
def polynomial_df(feature, degree):
    #initialize an empty dataframe
    poly_df = pd.DataFrame()
    poly_df['power_1'] = feature.values.tolist() #clumn power_1 contains feature values
    # first check if degree > 1
    if degree > 1:
        # then loop over the remaining degrees:
        # range usually starts at 0 and stops at the endpoint-1. We want it to start at 2 and stop at degree
        for power in range(2, degree+1):
            # first we'll give the column a name:
            name = 'power_' + str(power)
            poly_df[name] = feature.apply(lambda x: x**power)
    return poly_df
```

```python
df = data_train[['sqft_living','price']]
```

```python
poly_data = polynomial_df(df['sqft_living'], 15)
l2_small_penalty = 1e-5
```

```python
#importing ridge from sklearn linear model
from sklearn.linear_model import Ridge
```

```python
clf = Ridge(alpha = l2_small_penalty, normalize = True, fit_intercept = True)
x = poly_data
y = df['price']
ans = clf.fit(x,y)
print(pd.Series(ans.coef_, index = x.columns))
```

```python
def penalty(set_, penalty_value, power):
    poly_data = polynomial_df(set_['sqft_living'], power)
    l2_small_penalty = penalty_value
    clf = Ridge(alpha = l2_small_penalty, normalize = True, fit_intercept = True)
    x = poly_data
    y = set_['price']
    ans = clf.fit(x,y)
    results = pd.Series(ans.coef_, index = x.columns)
    y_predict = ans.predict(x)
    plt.plot(poly_data['power_1'],y,'.',
        poly_data['power_1'],y_predict,'-')
    return(results[0])
```
